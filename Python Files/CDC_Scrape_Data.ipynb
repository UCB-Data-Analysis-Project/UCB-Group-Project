```python
# Import Libaries

import requests
import json
import pandas as pd

#Create function to pull JSON data from CDC API

def CDC_Pull (ThemeId):
    url = "https://nccd.cdc.gov/DHDSPAtlas/proxy.ashx?https://proxy/api/data"
    payload = json.dumps({
      "CountyVersion": "2018",
      "GeographyType": "county",
      "ParentGeographyFilter": "",
      "ThemeClassId": 2,
      "ThemeFilterOptions": [],
      "ThemeId": ThemeId,
      "ThemeSubClassId": 13
    }, indent=4)
    headers = {
      'Content-Type': 'application/json',
      'Cookie': 'TS01ae363a=015d0abe8785e5b669589a99d047712fb1a00a60e1bccd9ee6909e7b2919ccad01d4b5ace6a65d1b843858eac8ee73347cee59aaf5'
    }
    response = requests.request("POST", url, headers=headers, data=payload)
    return response

# Request for the Percent of Black Population by County

Black_Pop_TID = 38

#Ensure correct JSON formatting

res = CDC_Pull(Black_Pop_TID).json()

# Convert RAW JSON into DataFrame

Json_DF_B = pd.json_normalize(res['Data'])

# Rename Value to Correct Column title, will change for every COl
## Display DataFrame

json_df_Black_POP = Json_DF_B.rename(columns={'Value': 'Percent of Black Population'})
json_df_Black_POP
 
```
