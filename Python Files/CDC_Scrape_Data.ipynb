```python
import requests
import json
import pandas as pd

url = "https://nccd.cdc.gov/DHDSPAtlas/proxy.ashx?https://proxy/api/data"

payload = json.dumps({
  "CountyVersion": "2018",
  "GeographyType": "county",
  "ParentGeographyFilter": "",
  "ThemeClassId": 2,
  "ThemeFilterOptions": [],
  "ThemeId": 38,
  "ThemeSubClassId": 13
}, indent=4)
headers = {
  'Content-Type': 'application/json',
  'Cookie': 'TS01ae363a=015d0abe8785e5b669589a99d047712fb1a00a60e1bccd9ee6909e7b2919ccad01d4b5ace6a65d1b843858eac8ee73347cee59aaf5'
}

response = requests.request("POST", url, headers=headers, data=payload)

res = response.json()

print(response.json())


```

```python
Json_DF = pd.json_normalize(res['Data'])
Json_DF
```

```python
json_df = Json_DF.rename(columns={'Value': 'Percent of Black Population'})
json_df
```
